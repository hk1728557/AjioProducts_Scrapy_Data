```python
items = []
for url in tqdm(urls):
    purl = "https://www.ajio.com" + (url)
    #print(purl)
    res = requests.get(purl)
    sel = Selector(res)
    text = res.content
    j_text = sel.xpath('//*/script[3]/text()').extract()
    j_text = sel.xpath('//*/script[3]/text()').extract()[0]
    j_text = j_text.split("\r\n          window.__PRELOADED_STATE__ = ")[1].replace("};\r\n","}").strip()
    j_text = json.loads(j_text)
    #j_text
    
    with open("pro_file.json","w") as f:
        f.write(json.dumps(j_text))
        f.close()
    
    
    with open("pro_file.json") as access_json:
        pread_content = json.load(access_json)
        
    item = {}
    item['product_url'] = purl   
    item['title_name'] = nested_lookup('name',nested_lookup('rilfnlBreadCrumb',pread_content))
    item['mrp'] = [*set(nested_lookup('value',nested_lookup('wasPriceData',nested_lookup('options',pread_content))))]
    item['offer_price'] = [*set(nested_lookup('value',nested_lookup('price',pread_content)))]
    #now set use to delete duplicate datas
    
    
    pro_DetailsKeys = nested_lookup('key',nested_lookup('mandatoryInfo',pread_content))
    pro_DetailsValues = nested_lookup('title',nested_lookup('mandatoryInfo',pread_content))
    pro_Details={}
    for pro_DetailsKey,pro_DetailsValue in zip(pro_DetailsKeys,pro_DetailsValues):
        pro_Details[pro_DetailsKey] = pro_DetailsValue
#     print(pro_Details)    
    
    item['Product_Code'] = nested_lookup('code',nested_lookup('variantOptions',pread_content))
    item['hsn_Code'] = nested_lookup('hsnCode',nested_lookup('variantOptions',pread_content))
    
    media_url={}
    img_urls = nested_lookup('url',nested_lookup('images',pread_content))
    img_urls = [*set(img_urls)] # delete duplicate data use @set
    for idx,img_url in  enumerate(img_urls):
        i_url ='img_url{}'
        img_url = [i_url.format(idx)]
        #print(img_url)

        for url_key,img_url in zip(img_url,img_urls):
            media_url[url_key] = "".join(img_url)

    #print(media_url)        

    
    
    
    
    
    items.append(item)
    item.update(pro_Details)
    item.update(media_url)
    
    
    
```
